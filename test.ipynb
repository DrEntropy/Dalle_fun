```python
import openai
from PIL import Image
import os
import requests
from credentials import api_key
openai.api_key = api_key
```

```python
# Create images directory if it doesnt exist
if not os.path.exists('images'):
    os.makedirs('images')
```

```python
response = openai.Image.create(
  prompt="charcoal sketch of a black and white tuxedo cat sitting on a chair in an empty room",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']
```

```python


im = Image.open(requests.get(image_url, stream=True).raw)
display(im)
```

```python
# save image
im.save('images/test.png')
```
