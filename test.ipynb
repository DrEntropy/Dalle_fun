```python
import openai
from PIL import Image
import os
import requests
from credentials import api_key
openai.api_key = api_key
```

```python
# Create images directory if it doesnt exist
if not os.path.exists('images'):
    os.makedirs('images')
```

```python
def gen_and_save(description, filename = None, style = "watercolor"):
  response = openai.Image.create(
  prompt = f"{style} of {description} \n",
  n = 1,
  size="512x512"
  )
  image_url = response['data'][0]['url']
  img = Image.open(requests.get(image_url, stream=True).raw)
  if(filename):
    img.save(f"images/{filename}")
  display(img)
  
```

```python
gen_and_save("A lighthouse that has fallen dark, on a rocky island","island.png")
```

```python
gen_and_save("A quiet beach with crashing waves and a rocky outcropping. ", filename="seascape.png")
```

```python
gen_and_save("" +
              "silhouette of a man standing on the path with glowing fish-like eyes")
```

```python
gen_and_save("a gold coin with an egyptian obelisk on it", style = "photograph")
```

```python
gen_and_save("Four humanoids with green skin and the head of a frog and  sharp claws" + \
             "coming forward as a group through a thicket", filename="frogman.png")
```
