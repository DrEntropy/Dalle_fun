```python
import openai
from PIL import Image
import os
import requests
from credentials import api_key
openai.api_key = api_key
```

```python
# Create images directory if it doesnt exist
if not os.path.exists('images'):
    os.makedirs('images')
```

```python
def gen_and_save(description, filename = "temp.png"):
  response = openai.Image.create(
  prompt = f"charcoal sketch of {description} \n",
  n = 1,
  size="1024x1024"
  )
  image_url = response['data'][0]['url']
  img = Image.open(requests.get(image_url, stream=True).raw)
  img.save(f"images/{filename}")
  display(img)
  
```

```python
gen_and_save("a room with two doors on different walls and a dinning room table","room.png")
```
